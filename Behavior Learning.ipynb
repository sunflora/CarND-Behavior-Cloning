```python
import pandas as pd
from sklearn.model_selection import train_test_split
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from keras.models import Sequential
from keras.models import Sequential
from keras.layers.core import Dense, Activation, Flatten, Dropout
from keras.layers.convolutional import Convolution2D
from keras.layers.pooling import MaxPooling2D
import json
from pathlib import Path
import os
```

```python
#Normalize image into range: -0.5 to 0.5
def normalize(image):
    image = image / 127 - 1.
    return image
```

```python
train_image_shape = (40, 160, 3)

sample_image = False

def display_sample_image():
    global sample_image
    
    filename = "dataUdacity/IMG/center_2016_12_01_13_30_48_287.jpg"
    image = cv2.imread(filename)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

    print("shape of image: ", image.shape)

    plt.figure()
    plt.imshow(image)
 
    plt.figure()
    image = image[55:image.shape[0]-25, 0:image.shape[1], :]
    print("shape of image: ", image.shape)
    plt.imshow(image)

    plt.figure()
    image = cv2.resize(image, (train_image_shape[1], train_image_shape[0]), interpolation=cv2.INTER_AREA)
    plt.imshow(image)

    plt.figure()
    image = cv2.cvtColor(image,cv2.COLOR_RGB2YUV)
    plt.imshow(image)
    
    plt.figure()
    image = normalize(image)
    plt.imshow(image)
    
    sample_image = image
     
```

```python
display_sample_image()
```

```python
def get_flipped_image(image, steering):
    flipped_image = cv2.flip(image, 1)
    flipped_steering = steering * -1.
    return flipped_image, flipped_steering

fi, fs = get_flipped_image(sample_image, 1)
print("flipped steering is ", str(fs))
plt.figure()
plt.subplot(1,4,1)
plt.imshow(sample_image)
plt.subplot(1,4,2)
plt.imshow(fi)
```

```python
def display_right_left_image():
    
    filename = "dataUdacity/IMG/center_2016_12_01_13_30_48_287.jpg"
    center_image = cv2.imread(filename)
    center_image = cv2.cvtColor(center_image,cv2.COLOR_BGR2RGB)

    filename = "dataUdacity/IMG/right_2016_12_01_13_30_48_287.jpg"
    right_image = cv2.imread(filename)
    right_image = cv2.cvtColor(right_image,cv2.COLOR_BGR2RGB)
    
    filename = "dataUdacity/IMG/left_2016_12_01_13_30_48_287.jpg"
    left_image = cv2.imread(filename)
    left_image = cv2.cvtColor(left_image,cv2.COLOR_BGR2RGB)
    
    plt.figure()
    plt.subplot(1,3,1)
    plt.imshow(left_image)
 
    plt.subplot(1,3,2)
    plt.imshow(center_image)

    plt.subplot(1,3,3)
    plt.imshow(right_image)
```
